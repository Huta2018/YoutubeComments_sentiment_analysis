```python
# Drop columns with object data types
df_cleaned = final_df.drop(columns=final_df.select_dtypes(include=['object']).columns)
df_cleaned.shape
```

```python
#Check  the presence of missing values
df_cleaned.isna().sum()
```

```python
#Handeling missing values
#missing values can be eiter filled with mean, median, or mode
#Here I will use mean
df = df_cleaned.fillna(df_cleaned.mean()).copy()

from sklearn.preprocessing import MinMaxScaler

# Initialize the scaler
scaler = MinMaxScaler()

# Fit and transform the data
df_normalized = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_result = pca.fit_transform(df_normalized)

plt.scatter(pca_result[:,0], pca_result[:,1])
plt.title('PCA Result')
plt.show()
```

```python
from sklearn.cluster import KMeans
import seaborn as sns

kmeans = KMeans(n_clusters=4)
df['cluster'] = kmeans.fit_predict(df_normalized)

sns.scatterplot(x=pca_result[:,0], y=pca_result[:,1], hue=df['cluster'])
plt.title('K-Means Clustering')
plt.show()
```

```python
from sklearn.neighbors import LocalOutlierFactor

lof = LocalOutlierFactor()
outliers = lof.fit_predict(df_normalized)

plt.scatter(df_normalized.iloc[:,0], df_normalized.iloc[:,1], c=outliers)
plt.title('Outlier Detection')
plt.show()
```
